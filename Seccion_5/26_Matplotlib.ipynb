```python
#libreria que permite graficar
import matplotlib.pyplot as plt
```

```python
#para graficar, se debe enviar un arreglo o alguna funcion
plt.plot([1,2,3,4])
plt.ylabel('grafico')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') #'ro' significa el color
plt.axis([0,6,0,20]) #Con esto definimos el tamaño de nuestros ejes
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(0.,5.,0.2)
t
```

```python
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g') #luego de cada color se puede enviar otro punto para graficar sobre el mismo plano
```

```python
def grafico(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.2)

plt.figure(1) # Pra determinar el primer plano del segundo plano
plt.subplot(211) #Subgrafico
plt.plot(t1,grafico(t1),'bo',t2,grafico(t2),'k')

plt.subplot(212) #Subgrafico
plt.plot(t2,np.cos(2*np.pi*t2),'--r')
plt.show()


```

```python
#para crear etiqueta en matplotlib
x = np.linspace(0,2,100) #Se crea un arreglo 
fig,ax=plt.subplots()
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label='quadratic')
ax.plot(x,x**3,label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple plot')
ax.legend()
```
